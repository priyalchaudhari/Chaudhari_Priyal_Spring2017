```python
import os
from os import listdir
from os.path import isfile, join
import glob
from email.parser import Parser
import email
```

```python
folder_dist={}
mypath = os.path.expanduser('~\\midterm\\data\\enron\\maildir')
employe_names = [x for x in os.listdir(mypath)]
for name in employe_names:
    newpath = os.path.expanduser('~\\midterm\\data\\enron\\maildir\\'+str(name))
    employee_folder_details = [x for x in os.listdir(newpath)]
    for folder in employee_folder_details:
        if name in folder_dist:
            folder_dist[name].append(folder)
        else:
            folder_dist[name] = [folder]
```
