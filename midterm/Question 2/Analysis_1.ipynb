# Opinion mining of Newspaper headlines using NLTK:
- Determining positive and negative score for each headkine and then calculating total score for whole month. 

```python
import os
from os import listdir
from os.path import isfile, join
import glob
import nltk
from nltk.corpus import sentiwordnet as swn
from nltk.corpus import wordnet as wn
import string
import operator
from glob import glob
import json
from matplotlib import pyplot as plt
```

```python
homedir = os.path.expanduser("~")
path_to_archive = homedir+"\\midterm\\data\\Archive API\\"
print(path_to_archive)
```

# Function for algorithm explained in ReadME file 
- here i have printed the result and graph for year 2015 using Archive API 
- only used 1 year because there is too much data and it takes time to run Hence ploted the grath for 2015 If want change the value for year and other output can be seen 

```python
for year in glob(path_to_archive+'\\*'):
    flag=True
    for subdirs, dirs, files in os.walk(year):
        score={}
        score_list=[]
        month_list=[]
        for file in files:
                headline_polarity={}
                negative_polarity_value=positive_polarity_value=0
                #try:
                with open(os.path.join(subdirs, file),'r', errors = 'ignore') as response:    
                    json_txt = json.load(response)
                    #print(json_txt)
                    for iter1 in range(len(json_txt["response"]["docs"])):
                        headline=(json_txt["response"]["docs"][iter1]["headline"]["main"]) # getting the headlines from the file 
                        #print(headline)
                        txt = nltk.word_tokenize(headline)  # tokanizing the words 
                        posTag=nltk.pos_tag(txt)         # pos_tagging the words 
                        lematizer = nltk.wordnet.WordNetLemmatizer()          
                        porter = nltk.PorterStemmer()                        #importing stemmer porter
                        n=p=0
                        for word in posTag:
                            lemma= lematizer.lemmatize(word[0])         #passing the words to lemmatizer 
                            stem = porter.stem(lemma)                    #then normalizing the words using stemmer 
                            net=wn.synsets(stem)                   # finding the syn set of the word
                            for i in net:    
                                breakdown = swn.senti_synset(i.name())   #passing those wods to sentisynset method to calculate the score of the wor
                                #print(breakdown," ",breakdown.obj_score())
                                n+=breakdown.neg_score()
                                #print(breakdown.neg_score()) # this method will print negative score 
                                p+=breakdown.pos_score()     # this method will give positive score 
                                #print(breakdown.pos_score())
                        if n>p:
                            headline_polarity[headline]=-1      # assigning the polarity based on scores 
                            negative_polarity_value -=1
                            #print('negative',negative_polarity_value)
                        elif n<p:
                            headline_polarity[headline]=1
                            positive_polarity_value +=1
                            #print('positive',positive_polarity_value)
                        else:
                            headline_polarity[headline]=0

                
                score1 = positive_polarity_value - negative_polarity_value  # calculating the total score 
                # plotting the values according to months. 
                
                f = file.split("_")
                f1 = f[1].split(".")
                if f[0] == "2015":
                    #plt.plot(score,f1[0])
                    score_list.append(score1)
                    month_list.append(f1[0])
                    score[int(f1[0])]=score1
                    flag=False
                print("Score for month",f1[0],"of year",f[0]," is :",positive_polarity_value - negative_polarity_value)
        if not flag:
            #sorted_characters = sorted(score.items(), key=operator.itemgetter(1), reverse = True) # sorting the dict based on frequency
            score_list=[]
            month_list=[]
            for i in range(1,13):
                score_list.append(score[i])
                month_list.append(i)
            #print(score_list," ",month_list)
            xtic=['Jan','Feb','March','Apr','May','Jun','July','Aug','Sep','Nov','Dec']
            plt.xticks(month_list,xtic)
            plt.xlabel("Month")
            plt.ylabel(" Total Sentiment score")
            plt.title(" Month wise Sentiment score")
            plt.plot(month_list,score_list)
            plt.show()
            flag=True
                   
```
